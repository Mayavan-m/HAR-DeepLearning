# Notebook 6: Results Comparison

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
models = ['CNN', 'LSTM', 'CNN-LSTM']
accuracy = [0.29, 0.93, 0.88]
precision = [0.41, 0.93, 0.88]
recall = [0.23, 0.93, 0.88]
f1 = [0.18, 0.93, 0.88]
```

```python
print(f"{'Model':<12} {'Accuracy':<12} {'Precision':<12}{'Recall':<12} {'F1':<12}")
print("-" * 60)
for i in range(len(models)):
    print(f"{models[i]:<12} {accuracy[i]:<12} {precision[i]:<12}{recall[i]:<12} {f1[i]:<12}")
    
```

```python
x = range(len(models))
width = 0.2
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar([i - 0.3 for i in x], accuracy, width, label='Accuracy')
ax.bar([i - 0.1 for i in x], precision, width, label='Precision')
ax.bar([i + 0.1 for i in x], recall, width, label='Recall')
ax.bar([i + 0.3 for i in x], f1, width, label='F1')
ax.set_ylabel('Score')
ax.set_title('Model Comparison: CNN vs LSTM vs CNN-LSTM')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()
plt.tight_layout()
plt.show()
```
