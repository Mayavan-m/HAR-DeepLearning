# Notebook 2: Prepare Data for Deep Learning

```python
import numpy as np
import os
from sklearn.preprocessing import StandardScaler
```

```python
base = r'C:\Users\M Mayavan\OneDrive\Desktop\HAR_Project\UCI HAR Dataset'
train_path = base + r'\train\Inertial Signals'
test_path = base + r'\test\Inertial Signals'
```

```python
signals = ['body_acc_x', 'body_acc_y', 'body_acc_z', 'body_gyro_x', 'body_gyro_y', 'body_gyro_z', 'total_acc_x', 'total_acc_y', 'total_acc_z']
```

```python
train_list = []
for s in signals:
    f = os.path.join(train_path, s + '_train.txt')
    data = np.loadtxt(f)
    train_list.append(data)
X_train = np.stack(train_list, axis=2)
print(X_train.shape)
```

```python
test_list = []
for s in signals:
    f = os.path.join(test_path, s + '_test.txt')
    data = np.loadtxt(f)
    test_list.append(data)
X_test = np.stack(test_list, axis=2)
print(X_test.shape)
```

```python
y_train = np.loadtxt(base + r'\train\y_train.txt')
y_test = np.loadtxt(base + r'\test\y_test.txt')
print(y_train.shape)
print(y_test.shape)
```

```python
scaler = StandardScaler()
X_train_flat = X_train.reshape(-1, 9)
X_test_flat = X_test.reshape(-1, 9)
scaler.fit(X_train_flat)
X_train = scaler.transform(X_train_flat).reshape(-1, 128, 9)
X_test = scaler.transform(X_test_flat).reshape(-1, 128, 9)
```

```python
y_train = y_train - 1
y_test = y_test - 1
```

```python
np.save(base + r'\X_train.npy', X_train)
np.save(base + r'\X_test.npy', X_test)
np.save(base + r'\y_train.npy', y_train)
np.save(base + r'\y_test.npy', y_test)
print('Data saved!')
```
